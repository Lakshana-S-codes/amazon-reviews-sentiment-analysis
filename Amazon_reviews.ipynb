```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count of each review rating
rating_counts = df['overall'].value_counts().sort_index()

# Plotting the distribution of review ratings
rating_counts.plot(kind='bar', colormap='viridis', figsize=(8,5))
plt.xlabel("Review Rating (Stars)")
plt.ylabel("Count")
plt.title("Distribution of Amazon Review Ratings")
plt.show()

print(rating_counts)
```

```python
# Conversion factor: 1 week = 7 days
def convert_days_to_weeks(days):
    return days / 7

# Applying conversion to day_diff column
df['week_diff'] = df['day_diff'].apply(convert_days_to_weeks)

# Displaying the first few rows
df[['day_diff', 'week_diff']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot of day_diff vs. helpful_yes, colored by overall rating
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['day_diff'], y=df['helpful_yes'], hue=df['overall'], alpha=0.7, palette='viridis')
plt.xlabel('Days Since Review')
plt.ylabel('Helpful Votes')
plt.title('Helpfulness of Reviews Over Time')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Create the heatmap
fig, ax = plt.subplots(figsize=(10, 6))
cax = ax.matshow(corr_matrix, cmap="coolwarm")

# Add color bar
plt.colorbar(cax)

# Set axis ticks and labels
ax.set_xticks(np.arange(len(corr_matrix.columns)))
ax.set_yticks(np.arange(len(corr_matrix.columns)))
ax.set_xticklabels(corr_matrix.columns, rotation=90)
ax.set_yticklabels(corr_matrix.columns)

# Add title
plt.title("Correlation Heatmap of Review Metrics", pad=20)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create vote bins
df['vote_group'] = pd.cut(df['total_vote'], bins=[-1, 0, 5, 20, df['total_vote'].max()],
                          labels=['None', 'Low', 'Medium', 'High'])

plt.figure(figsize=(8, 5))
sns.boxplot(x='vote_group', y='wilson_lower_bound', data=df, palette='coolwarm')
plt.xlabel('Total Vote Activity Level')
plt.ylabel('Wilson Lower Bound Score')
plt.title('WLB Score by Vote Activity Level')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by day_diff to count number of reviews
review_trend = df.groupby('day_diff')['reviewText'].count().reset_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x='day_diff', y='reviewText', data=review_trend, marker='o', color='crimson')
plt.xlabel("Days Since Review")
plt.ylabel("Number of Reviews")
plt.title("Trend of Review Counts Over Time")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Strip column names just in case
df.columns = df.columns.str.strip()

# Bin vote-related columns
df['vote_group'] = pd.cut(df['total_vote'], bins=[-1, 0, 5, 20, df['total_vote'].max()],
                          labels=['None', 'Low', 'Medium', 'High'])
df['helpful_group'] = pd.cut(df['helpful_yes'], bins=[-1, 0, 3, 10, df['helpful_yes'].max()],
                             labels=['None', 'Some', 'Helpful', 'Very Helpful'])

# Plot
sns.relplot(x='day_diff', y='overall', hue='vote_group', style='helpful_group', data=df, kind='scatter')
plt.xlabel('Days Since Review')
plt.ylabel('Rating (Overall)')
plt.title('Review Age vs Rating by Vote & Helpfulness')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Sort values by review age
df_sorted = df.sort_values(by='day_diff')

x = df_sorted['day_diff']
y = df_sorted['overall']

sns.lineplot(x=x, y=y, color='red', marker='o')
plt.xlabel('Days Since Review')
plt.ylabel('Overall Rating')
plt.title('Rating Fluctuations Over Review Age')
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("amazon_reviews.csv")

# Display the first 10 rows of the dataset
df.head(10)

# Get dataset info including column names, data types, and missing values
df.info()

# Get summary statistics for numerical columns
df.describe()
```

```python
# Fill numeric columns with their mean
df.fillna(df.select_dtypes(include=['number']).mean(), inplace=True)

# Fill categorical columns with their mode
df.fillna(df.select_dtypes(include=['object']).mode().iloc[0], inplace=True)

# Display the first 10 rows to verify
print(df.head(10))
```
